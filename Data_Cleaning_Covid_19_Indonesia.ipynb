```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import plotly.graph_objects as go
```

```python
#menampilkan dataframe
data_covid = pd.read_csv('/content/drive/MyDrive/porto/covid_19_indonesia_time_series_all.csv')
data_covid.head()
```

##DATA PREPERATION

```python
data_covid.info()
```

```python
#mengubah data type pada feature Date menjadi datetime
import datetime as dt
NOW = dt.datetime(2020, 2, 2)
data_covid['Date'] = pd.to_datetime(data_covid['Date'])
```

```python
data_covid.info()
```

##DATA CLEANING

```python
#cek null dalam persen data
data_covid.isnull().mean()*100
```

```python
#cek null dalam persen data
data_covid.isna().mean()*100
```

```python
#drop kolom yang memiliki hasil persentasenya 100
data_covid = data_covid.drop('City or Regency', axis=1)
```

```python
#inspect feature'Special Status'
data_covid['Special Status'].value_counts()
```

```python
data_covid = data_covid.drop('Special Status', axis=1)
```

```python
#inspect feature 'Growth Factor of New Cases', 'Growth Factor of New Deaths'
data_covid[['Growth Factor of New Cases', 'Growth Factor of New Deaths']].head(10)
```

```python
#mengganti NaN pada feature 'Growth Factor of New Cases', 'Growth Factor of New Deaths' dengan 0
data_covid['Growth Factor of New Cases'] = data_covid['Growth Factor of New Cases'].fillna(float(0))
data_covid['Growth Factor of New Deaths'] = data_covid['Growth Factor of New Deaths'].fillna(float(0))
```

```python
#inspect feature 'Province', 'Island','Time Zone'
data_covid[['Province', 'Island','Time Zone']].head(10)
```

```python
#mengganti NaN of Province, Island, Time Zone menjadi Unknown
data_covid.update(data_covid[['Province', 'Island', 'Time Zone']].fillna('Unkown'))
```

```python
data_covid[['Province','Total Cities', 'Total Urban Villages','Total Rural Villages']].tail(50)
```

```python
##mengganti NaN pada feature 'Province','Total Cities', 'Total Urban Villages','Total Rural Villages' dengan 0
data_covid['Total Cities'] = data_covid['Total Cities'].fillna(float(0))
data_covid['Total Urban Villages'] = data_covid['Total Urban Villages'].fillna(float(0))
data_covid['Total Rural Villages'] = data_covid['Total Rural Villages'].fillna(float(0))
```

```python
#hasil dari data cleaning, bahwa tidak ditemukan feature yang kosong, null, atau NaN
data_covid.isna().mean()*100
```
